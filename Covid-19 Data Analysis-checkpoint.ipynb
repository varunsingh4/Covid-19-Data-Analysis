# Finding maximum infection rate for all countries


```python
country=list(corona_dataset_aggregated.index)
max_infection_rate=[]
for c in country:
    max_infection_rate.append(corona_dataset_aggregated.loc[c].diff().max()) #Append function adds a single item max_infection_rate list to the returned by the aggregrated dataset
max_infection_rate
corona_dataset_aggregated['Maximum_Infection_rate']=max_infection_rate
    
    
```
